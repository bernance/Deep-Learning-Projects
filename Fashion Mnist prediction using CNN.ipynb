```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers, datasets
```

```python
(X_train, y_train), (X_test, y_test) = datasets.fashion_mnist.load_data()
```

```python
X_test.shape
```

```python
print(X_train.shape)
```

## Creating an ANN model

```python
X_train = (X_train / 255)
X_test = (X_test / 255)
```

```python
model = keras.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(5000, activation='relu'),
    layers.Dense(1000, activation='relu'),
    layers.Dense(10, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```

## Integrating the CNN layer

### Reshaping the dataset

```python
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)
```

```python
model1 = keras.Sequential([
    #CNN
    layers.Conv2D(filters= 50, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPool2D((2,2)),

    layers.Conv2D(filters= 50, kernel_size=(3,3), activation='relu'),
    layers.MaxPool2D((2,2)),
    
    #Flatten layer
    layers.Flatten(),
    #Dense Layers
    layers.Dense(5000, activation='relu'),
    layers.Dense(1000, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model1.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model1.fit(X_train, y_train, epochs=5)
```

```python
model1.evaluate(X_test, y_test)
```

```python
model2 = keras.Sequential([
    #CNN
    layers.Conv2D(filters= 40, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPool2D((2,2)),

    layers.Conv2D(filters= 40, kernel_size=(3,3), activation='relu'),
    layers.MaxPool2D((2,2)),
    
    #Flatten layer
    layers.Flatten(),
    #Dense Layers
    layers.Dense(150, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(50, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(10, activation='softmax')
])
```

```python
model2.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model2.fit(X_train, y_train, epochs=10)
```

```python
model2.evaluate(X_test, y_test)
```
